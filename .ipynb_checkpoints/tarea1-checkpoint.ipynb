```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("T1_Housing.csv") 
X = data["Avg Area Income"]
Y = data["Price"]
```

```python
def plotData():
    plt.plot(X, Y, "*")
    plt.title("Ingreso promedio vs Precio")
    plt.xlabel("Ingreso promedio")
    plt.ylabel("Precio de las casas")
    plt.show()

def reg_lineal(X,Y,rho):
    """
    in:
    out:
    """
    X = np.array(X).reshape(len(X), 1)
    Y = np.array(Y).reshape(len(Y), 1)
    unos = np.ones(len(X)).reshape(len(X),1)
    xTilda = np.concatenate([X, unos], 1)
    I = np.identity(2)
    theta = np.linalg.inv(np.dot(xTilda.transpose(), xTilda) + rho * I)
    theta = np.dot(theta, np.dot(xTilda.transpose(), Y))
    return theta
    
def plotThetaRho():
    p = 10
    x = np.arange(p + 1)
    a = []
    b = []
    for rho in x:
        a.append(reg_lineal(X, Y, rho)[0])
        b.append(reg_lineal(X, Y, rho)[1])
    plt.plot(x, a, "*")
    plt.show()
    plt.plot(x, b, "*")
    plt.show()

def plotModelo(rho):
    a = reg_lineal(X, Y, rho)[0] #Optimizar
    b = reg_lineal(X, Y, rho)[1]
    #Prediciones:
    yGorro = []
    for x in X:
        yGorro.append(a*x + b)
    plt.plot(X, Y, "*")
    plt.plot(X, yGorro)
    plt.show()


```
